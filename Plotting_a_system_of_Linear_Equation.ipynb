```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 60, 1000)
```

**Distnce travelled by robber: d = 2.5t**

```python
d_r = 2.5 * t
```

**Distance travelled by Sherrif d = 3(t-5)**

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('Time taken to catch a bank roberr')
plt.xlabel('Time (in minutes)')
plt.ylabel('Distance (in KM)')
ax.set_xlim(0,40)
ax.set_ylim(0,  100)
ax.plot(t, d_r, color='violet')
ax.plot(t, d_s, color='green')
plt.axvline(x=30, color='crimson', linestyle='--')
plt.axhline(y=75, color='crimson', linestyle='--')
```
